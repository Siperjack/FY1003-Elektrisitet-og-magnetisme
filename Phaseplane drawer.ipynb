```python
import numpy as np
import matplotlib.pyplot as plt
```

### Function

```python
def drawer(f,sides = 10, M = 2000, xypos = False):
    if xypos:
        xs = np.linspace(0,sides, M)
    else:
        xs = np.linspace(-sides/2,sides/2, M)
    ys = xs
    ux,uy = np.meshgrid(xs,ys)
    grid = f(ux,uy)
    plt.figure(figsize=(8, 5), dpi=80)
    plt.streamplot(xs,ys,grid[0],grid[1])
    plt.show()
    #plt.streamplot(xs[1:M-1:M//20],ys[1:M-1:M//20],grid[0][::M//20,::M//20],grid[1][::M//20,::M//20])
```

```python
def f(x,y):
    return np.array([x+np.exp(-y),-y])
def vanderPol(x,y):
    return np.array([y,-x+y*(1-x**2)])
def dipolFixedPoint(x,y):
    return np.array([2*x*y,y**2-x**2])
def twoEyedMonster(x,y):
    return np.array([y+y**2,-0.5*x+0.2*y-x*y+1.2*y**2])
def parrot(x,y):
    return np.array([y+y**2,-x+0.2*y-x*y+1.2*y**2])
```

```python
drawer(f)
```

```python
drawer(vanderPol)
```

```python
drawer(dipolFixedPoint, sides= 1)
```

```python
drawer(twoEyedMonster,sides = 10)
```

```python
drawer(parrot)
```

```python
def f(x,y):
    return np.array([y**3-4*x,y**3 - y -3*x])
```

```python
drawer(f,sides = 40)
```

```python
def eks73(x,y,a = 0.08 , b = 0.6): 
    return np.array([-x + a*y + x**2*y, b - a*y -x**2*y])

drawer(eks73, sides = 3, xypos = True)
```

```python
def task4(x,y):
    return np.array([-4*x**3 - y, x**5 -2*y**3])
drawer(task4)
```
